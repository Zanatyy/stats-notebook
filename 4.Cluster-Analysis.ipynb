# Cluster Analysis

### Cluster Analysis

**Definition**: It's a form of exploratory data analysis technique where observations are divided into meaningful groups that share common characteristics and atrributes.

**Key Points:**

- **Purpose**: 
  - To identify natural groupings in data.
  - To simplify data by reducing the number of observations into meaningful clusters.
  - To discover patterns and insights that may not be immediately obvious.


- **Basic Approaches and Techniques**:

1- Centroid Clustering:


2- Density-based clustering:

3- Distribution Clustering:

4- Hierarchical Clustering:



- **Common Techniques**:
  - **K-Means Clustering**: Partitions data into \( $k$ \) clusters based on the mean of the points in each cluster. It's simple and widely used.
  - **Hierarchical Clustering**: Builds a tree of clusters by either a bottom-up (agglomerative) or top-down (divisive) approach, allowing for a more flexible structure.
  - **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**: Groups together points that are close to each other based on a distance measurement, allowing for clusters of varying shapes and sizes.

- **Distance Measures**: 
  - The similarity or dissimilarity between data points is often measured using metrics like Euclidean distance, Manhattan distance, or cosine similarity.

- **Choosing the Number of Clusters**: 
  - In methods like K-means, selecting the right number of clusters can be challenging. Techniques such as the Elbow Method, Silhouette Score, or Gap Statistic can help determine an optimal number.

- Example 1: Using $k$-mean clustering in synthetic/generated dataset in code:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Create a K-Means model
kmeans = KMeans(n_clusters=4)

# Fit the model to the data
kmeans.fit(X)

# Predict cluster labels
y_kmeans = kmeans.predict(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')  # Mark centroids
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

The x marks on the plot are the centroid of the clusters.

- Example 2: Using $k$-mean in The Wine dataset from Scikit-Learn

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the Wine dataset
wine = datasets.load_wine()
X = wine.data  # Use the features for clustering

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a K-Means model
kmeans = KMeans(n_clusters=3, random_state=0)

# Fit the model to the data
kmeans.fit(X_scaled)

# Predict cluster labels
y_kmeans = kmeans.predict(X_scaled)

# Plot the results (using the first two features for visualization)
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')  # Mark centroids
plt.title('K-Means Clustering on Wine Dataset')
plt.xlabel('Feature 1 (Alcohol)')
plt.ylabel('Feature 2 (Malic Acid)')
plt.show()
```
